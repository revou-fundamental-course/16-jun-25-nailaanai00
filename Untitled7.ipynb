```python
# Import library pandas
import pandas as pd

# Membaca file CSV
df = pd.read_csv("earthquake_data_tsunami.csv")

# Menampilkan 5 baris pertama sebelum manipulasi
print("=== Data Awal ===")
display(df.head())

# ----------------------------
# 1. Mengubah nama kolom
# -----------------------------
df.rename(columns={'Month': 'Bulan', 'depth': 'Kedalaman'}, inplace=True)
print("\n=== Data Setelah Diubah Nama Kolom ===")
display(df.head())

# -----------------------------
# 2. Mengurutkan berdasarkan kolom 'Kedalaman' (dari terkecil ke terbesar)
# -----------------------------
df.sort_values(by='Kedalaman', ascending=True, inplace=True)

print("\n=== Data Setelah Diurutkan Berdasarkan Kedalaman ===")
display(df.head())

# ----------------------------------------
# 3. Menghapus kolom yang tidak diperlukan
# ----------------------------------------
df.drop(columns=['gap'], inplace=True)
print("\n=== Data Setelah Menghapus Kolom Gap ===")
display(df.head())

# -----------------------------------
# 4. Menghapus baris pertama (index 0)
# -----------------------------------
df.drop(index=0, inplace=True)

print("\n=== Data Setelah Menghapus Baris Pertama ===")
display(df.head())

# -----------------------------
# 5. Mengurutkan index ulang (ascending)
# -----------------------------
df.sort_index(ascending=True, inplace=True)

print("\n=== Data Setelah Index Diurutkan Ulang ===")
display(df.head())

# -----------------------------
# 6. Menampilkan gempa dengan magnitudo di atas 7
# -----------------------------
df_magnitude_tinggi = df[df["magnitude"] > 7]

print("\n=== Gempa dengan Magnitudo di Atas 7 ===")
display(df_magnitude_tinggi)

# -----------------------------
# 7. Menampilkan gempa yang menyebabkan tsunami
# -----------------------------
df_tsunami = df[df["tsunami"] == 1]

print("\n=== Gempa yang Menyebabkan Tsunami ===")
display(df_tsunami)

# -----------------------------
# 8. Menampilkan gempa yang terjadi pada tahun 2022
# -----------------------------
df_2022 = df[df["Year"] == 2022]

print("\n=== Gempa yang Terjadi pada Tahun 2022 ===")
display(df_2022)
```
